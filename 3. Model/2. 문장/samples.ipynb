```python
dir = "./main"
dir_write = []
movie = []
```

```python
def search(dirname):
    try:
        filenames = os.listdir(dirname)
        for filename in filenames:
            full_filename = os.path.join(dirname, filename)
            if os.path.isdir(full_filename):
                search(full_filename)
            else:
                ext = os.path.splitext(full_filename)[-1]
                if ext == '.mp4': 
                    temp = re.findall(r'Pitchfork/([0-9/_]+.mp4)', full_filename)[0]
                    movie.append(full_filename)
                    dir_write.append(temp)
                    # dir_write.append(re.sub('\\\\', '/', temp))
    except PermissionError:
        pass
```

```python
def samples(movie, dir_write):
    for cnt in range(len(dir_write)):
        name = os.path.splitext(movie[cnt])
        txt = ''.join([name[0], '.txt'])
        l = []
        with open(txt) as file:
            lines = file.readlines()
            for line in lines:
                l.append(re.sub('[-=+#/?_:^$@*\"※~&%!.,\(\{\[\]\}\)]', '', line.strip()).upper())
            l = ' '.join(l)
        with open('./media/lrs2_test_samples.txt', 'a') as f:
            f.write(''.join([dir_write[cnt], ', ']))
            f.write(''.join([l, '\n']))
```

```python
samples(movie, dir_write)
```
