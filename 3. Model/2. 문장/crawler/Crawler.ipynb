```python
from pytube import YouTube
from pytube import Playlist
from youtube_transcript_api import YouTubeTranscriptApi
import json
import ffmpeg
import youtube_dl
import os
```

```python
fail = []
count = 113
pl = Playlist("https://www.youtube.com/playlist?list=PLrIyFAtNxiQZRuVlAPQbhTpyiO00vk8ww")
```

```python
fail = []
count = 1000
pl = Playlist("https://www.youtube.com/playlist?list=PLztAHXmlMZFS9ZN7GTlZ2UOB2JmxICdt8")
```

```python
def download_video(url):
    print(url)
    yt = YouTube(url)
    yt.streams.filter(progressive=True, file_extension='mp4', res="720p")[0].download(output_path='./main', filename=str(count).zfill(6)+'.mp4')
    transcript = YouTubeTranscriptApi.get_transcript(yt.video_id, languages={'en'})
    with open('./main/'+str(count).zfill(6)+'.txt', 'w', encoding='utf-8') as json_file:
        json.dump(transcript, json_file)
```

```python
pl.length
```

```python
for url in pl.video_urls:
    try:
        download_video(url)
    except:
        fail.append(url)
    count = count + 1
```
